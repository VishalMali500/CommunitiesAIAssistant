```python
from pinecone import Pinecone

pc = Pinecone(api_key="pcsk_3Vkjww_Kc5Y2pMZkGWDnvsvjjT1B7ZBUKvA4zr6ZBZeDmW4CGtTBmD9SzvdDo84xPLuh5k")
index = pc.Index("communitiesvectordb")
```

```python
import os

data_folder = os.path.join(os.getcwd(), "Data")

files = os.listdir(data_folder)
all_docs = []
for file in files:
    all_docs.append(TextLoader(file_path= data_folder + "\\" + file, encoding="utf-8").load())
```

```python
for x in all_docs:
    for y in x:
         y.metadata["source"] = y.metadata["source"].split("\\")[-1].split(".")[0]
```

```python
splitter  = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
splits =[]
for t in all_docs:
    splits.extend(splitter.split_documents(t))
```

```python
embedding = HuggingFaceEmbeddings(model = "all-MiniLM-L6-v2")
```

```python
db  = PineconeVectorStore(embedding=embedding, index= index )
db.add_documents(splits)
```

```python
db.similarity_search_with_score("What is the frustration of Chloe Zhao ?")
```
